## Question 1
* Cluster item lv into each portion (dataset is at model lv) (5 points)
* From the dataset 1 Please calculate and fill the data to this table

```python
# Merge platform data with pricing data
merged_data = pd.merge(pricing_project_df, platform_number_df, left_on='grass_region', right_on='region')
```

```python
# Calculate Order Coverage
merged_data['order_coverage'] = merged_data['shopee_order']/merged_data['platform order']
```

```python
# Calculate Net Competitiveness
merged_data['competitiveness'] = np.where(merged_data['shopee_model_competitiveness_status'] == 'Shopee < CPT', 1, 
                                          np.where(merged_data['shopee_model_competitiveness_status'] == 'Shopee > CPT', -1, 0))
```

```python
merged_data.head()
```

```python
# # Group by region and calculate metrics
result = merged_data.groupby('grass_region').agg({
    'order_coverage': 'sum', 
    'competitiveness': 'mean', 
    'shopee_item_id': 'nunique'
}).reset_index()
```

```python
# Rename columns
result.columns = ['Region', 'Order Coverage (by Item)', 'Net Competitiveness (by Item)', '# of Item']
```

## Question 2
* Please write a python or pyspark code to re-arrange the data set to item_lv with the given priority  (5 points)

```python
# Create a new column with the priority of the status
status_priority = {'Shopee < CPT': 0, 'Shopee = CPT': 1, 'Shopee > CPT': 2}
pricing_project_df['status_priority'] = pricing_project_df['shopee_model_competitiveness_status'].map(status_priority)
```

```python
pricing_project_df.head()
```

```python
# Group by 'shopee_item_id' and assign the lowest priority status to each item
item_level_df = pricing_project_df.groupby('shopee_item_id')['status_priority'].min().reset_index()
```

```python
# Map back to the original status
priority_status_reverse = {0: 'Shopee < CPT', 1: 'Shopee = CPT', 2: 'Shopee > CPT'}
item_level_df['item_competitiveness_status'] = item_level_df['status_priority'].map(priority_status_reverse)
```

```python
# Drop the 'status_priority' column since it's no longer needed
item_level_df.drop(columns='status_priority', inplace=True)
```

```python
# Rename shopee_item_id to item_id
item_level_df.rename(columns={'shopee_item_id': 'item_id'}, inplace=True)
```

```python
# Display the new item level dataframe
item_level_df
```

## Question 3
* How many item at top 30% of model that contribute in highest order to platform by each country/region? (5 points)

```python
# Sort the dataset by 'shopee_order' in descending order to identify the top contributors
df_sorted = pricing_project_df.sort_values(by='shopee_order', ascending=False)
```

```python
df_sorted
```

```python
# Calculate the top 30% of items
top_30_percent_count = int(len(df_sorted)*0.3)
top_30_percent_count
```

```python
# Select the top 30% of items
df_top_30_percent = df_sorted.head(top_30_percent_count)
```

```python
# Count the number of items contributing to the highest orders by each country/region
top_item_by_region = df_top_30_percent.groupby('grass_region')['shopee_order'].count()
```

```python
# Display the results
top_item_by_region = top_item_by_region.reset_index(name='# of items')
top_item_by_region.rename(columns={'grass_region': 'Region'}, inplace=True)

top_item_by_region
```
